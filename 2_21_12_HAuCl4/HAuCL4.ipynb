# Absorbance spectra from 1mM HAuCl4

### To try to confer that all Au is consumed in NP process

```python
from pyuvvis import *
from pyuvvis.IO.gwu_interfaces import from_oceanoptics
plt.figsize(7,7)

ts=from_oceanoptics('HAuCl4_data', sample_by=2)
ts.name='HAuCl4_data'
ts.save('HAuCl4.pickle')

ts=ts.sub(ts.darkseries, axis=0)

specplot(ts)
ts=ts.as_interval('s')
areaplot(ts)
```

```python
h20=ts.ix[:,100.0:200.0]
mid=ts.ix[:,402.0:450.0]
end=ts.ix[:,500.::]
```

# Absorbance HAuCL4

```python
end.baseline=h20.mean(axis=1)

absplot(end)
```

## Interesting transient event directly after drop that was not stirring

```python
mid.baseline=0
specplot(mid)
absplot(mid)
```

# Conclusions

# Absorbance is huge, in fact saturated, by this mere 1mM concentration.  Therefore, even small amount in gold should be highly noticable.  Interesting peak at about 270nm disappears and becomes weight at 320nm.
