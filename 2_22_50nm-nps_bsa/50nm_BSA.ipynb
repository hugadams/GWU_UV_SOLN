```python
from pyuvvis import specplot, absplot, areaplot, TimeSpectra
from pyuvvis.IO.csv_wrapper import df_from_directory
from pyuvvis.nptools import haiss_m1, haiss_m2, haiss_m3, haiss_conc
plt.figsize(8,8)
```

### DATA DID NOT SAVE HEADERS SO I CANT USE FROM_OCEAN_OPTICS IMPORT.  INSTEAD WILL CONSTRUCT A DATAFRAME AND MANUALLY TURN IT INTO A TIMESPECTRA BY SETTING A SPECTRAL INDEX, DARK SERIES AND DEFINING AN ARBITRARY TIME PERIOD (TO ALLOW FOR AREAPLOTS AND OTHER TIME-DEPENDENT PLOTS).

```python
df=df_from_directory('./data', {'sep':'\t'})
dark=df.pop('100_40_10_dark.txt')
ts=TimeSpectra(df, specunit='nm')

ts.set_specindex(200.0, 850.0)
ts.set_daterange(start='2/22/12', periods=len(ts.columns), freq='4s')
ts.darkseries=dark

### Output for later use
ts.save('50nmnps_bsa.pickle')
```

```python
ts=ts.sub(ts.darkseries, axis=0)
ts=ts.as_interval('s')
specplot(ts)
areaplot(ts)
```

```python
h20=ts.ix[:,50.:100.]
nps=ts.ix[:,400.:600.]
bsa1=ts.ix[:, 900.:1050.]
bsa_drop2=ts.ix[:, 1250.:1400.]
```

```python
nps.baseline=h20.mean(axis=1)
bsa1.baseline=h20.mean(axis=1)

absplot(nps, color='g')

### Add other bsa plots manually
bsa1.iunit='a'
plt.plot(ts.index, bsa1, color='r')
plt.xlim(350.,850.)
plt.ylim(0.0, 0.6)
```

**This corresponds to a peak shift of 4.82nm** as show in the code cell below:


```python
spr_np=nps.ix[400.0:800.0].as_iunit('a').idxmax().mean()
spr_bsa1=bsa1.ix[400.0:800.0].as_iunit('a').idxmax().mean()

print 'SPR max in h20: %snm.  Sprtmax after BSA: %snm.  Delta %snm'%(spr_np, spr_bsa1, spr_bsa1-spr_np)
```

Quantify change due to bsa by rereferencing drop1 relative to nps.

```python
bsa1.iunit=None
nps.iunit=None

npbaseline=nps.mean(axis=1)

bsa1.baseline=npbaseline
x=absplot(bsa1)
plt.xlim(400.,800.)
plt.ylim(0.0, 0.085)
```
