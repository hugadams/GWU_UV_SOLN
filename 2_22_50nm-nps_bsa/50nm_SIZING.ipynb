```python
plt.figsize(8,8)

from pyuvvis import *
ts=tsload('50nmnps_bsa.pickle')
ts=ts.sub(ts.darkseries, axis=0)
ts.specunit='nm'
specplot(ts)
```

```python
ts=ts.as_interval('s')
areaplot(ts)
```

```python
h20=ts.ix[:,50.0:100.0]
nps=ts.ix[:,300.0:500.0]

nps.baseline=h20.mean(axis=1)
absplot(nps)
```

## Looking more closely at the peak value

```python
## define a quick rounding function for nice output
r1=lambda x: round(x,1)

nps.iunit='a'
sprmax=nps.ix[400.0:800.0].idxmax(axis=0).mean()
print 'spr_max at lambda = %s nm'%r1(sprmax)
```

```python
absplot(nps.ix[527.:530.])
```

### In reality, spr_max is probably closest to 528.5nm!  Tannic acid may also affect size estimation.

## Notice the tannic acid UV signature and how it varies from citrate!

# HAISS M1 (This should be pretty accurate for these large particles)

```python
from pyuvvis.nptools import *
    
print 'd = %s nm (Haiss method 1)'%r1(haiss_m1(nps).mean())
print 'SURPRISINGLY LOW'
```

## Haiss m2 (Pretty big discrepancy here based on theoretical and experimental fits)

```python
print 'd = %s nm (Haiss method 2 experimental fit)'%r1(haiss_m2(nps).mean())
print 'd = %s nm (Haiss method 2 theoretical fit)'%r1(haiss_m2(nps, exp=False).mean())
```
