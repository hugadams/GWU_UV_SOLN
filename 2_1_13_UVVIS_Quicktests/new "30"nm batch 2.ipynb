# Experimental results in centroid measurements from sem were approx 22nm; let's check the haiss function using both the experimental/theoretical fits and using 450nm and 430nm.

Haiss M1 (should not be extremely accurate for $d<25$ nm nps).  Will try with and without smoothing.

```python
from pyuvvis.nptools.haiss import haiss_m1, haiss_m2, haiss_m3, haiss_conc

## define a quick rounding function for nice output
r1=lambda x: round(x,1)

print 'd = %s nm (Haiss m1 with 2.0nm smoothing)'%r1(haiss_m1(nps, width=2.0).mean())
```

```python
print 'd = %s nm (Haiss method1) with NO smoothing'%r1(haiss_m1(nps, width=None).mean())
```

We see a 3nm diameter estimate discrepancy due to uncertainty in the peak wavelength when smoothed or unsmoothed.  Let's take a look at the smoothed and unsmoothed data more closely using a boxcar width of 2nm.

```python
absplot(nps.ix[520.:530.].boxcar(2.0), title='Smoothing')
absplot(nps.ix[520.:530.], title='No smoothing')
```

We see here that smoothing actually hurts the dataset and so are content with non-smoothing results which predict **d=31.9nm** (however this method is perhaps not valid).
