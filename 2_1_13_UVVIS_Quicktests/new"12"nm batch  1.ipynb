```python
from pyuvvis.IO.gwu_interfaces import from_oceanoptics
plt.figsize(8,8)

ts=from_oceanoptics('12nm_batch1')
ts.name='12nm_batch1' #will go to plot titles

from pyuvvis import specplot
specplot(ts)
```

!!Don't forget to subtract the darkseries before doing any absorbance data!!

```python
ts=ts.sub(ts.darkseries, axis=0) #axis=0 is important
specplot(ts)
```

```python
from pyuvvis.pyplots.basic_plots import areaplot
ts=ts.as_interval('m')
areaplot(ts)
```

```python
h20=ts.ix[:, 0.1:0.3]
nps=ts.ix[:, 0.8:1.5]

nps.baseline=h20.mean(axis=1)

from pyuvvis import absplot
absplot(nps)
```

```python
nps.iunit='a'
(nps.ix[410.0:450.0].idxmin(axis=0)).mean()
absplot(nps.ix[410.0:450.0])
```

```python
nps.ix[410.0:450.0].idxmin(axis=0)
```

```python
from pyuvvis.nptools.haiss import haiss_m1, haiss_m2

## define a quick rounding function for nice output
r1=lambda x: round(x,1)

print 'd = %s nm (Haiss m1 with 2.0nm smoothing)'%r1(haiss_m1(nps, peak_window=2.0).mean())
```

```python
print 'd = %s nm (Haiss method1) with NO smoothing'%r1(haiss_m1(nps, width=None).mean())
```

```python
absplot(nps.ix[515.:525.].boxcar(2.0), title='Smoothing')
absplot(nps.ix[515.:525.], title='No smoothing')
```

```python
print 'd = %s nm  (Experimental parms, ref=450 nm)'%r1(haiss_m2(nps, ref=450.0,exp=True).mean())
print 'd = %s nm  (Theoretical parms, ref=450 nm)'%r1(haiss_m2(nps, ref=450.0, exp=False).mean())
print 'd = %s nm  (Experimental parms, ref=432 nm)'%r1(haiss_m2(nps, ref=432.0, exp=True).mean())
print 'd = %s nm  (Theoretical parms, ref=432 nm)'%r1(haiss_m2(nps, ref=432.0, exp=False).mean())
```


1. Smoothing data from the USB650 spectrometer is not a good idea because we only have nm resolution; whereas, the fibers using the usb2000 spectrometer have triple this resolution (eg 3 data points per nm).

2. The most accurate Haiss function is likely M2 with experimental parms and ref at 450nm; however, size prediction seems to range between d = 14.7nm and 14.9nm (if you trust that using the absorbance at 450nm is ok (will have to look back at haiss paper carefully.)

3. Centroid-centroid distances from a single fiber image place this value closer to 13.88nm.
