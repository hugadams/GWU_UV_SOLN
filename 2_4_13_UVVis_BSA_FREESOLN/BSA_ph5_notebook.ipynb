Let's load the full data, stored in the 'rundata.pickle' object which is the entire spectral dataset from 2/1/13 in which I added BSA to ph5 NP's and osbserved binding.

```python
from pyuvvis import *
plt.figsize(8,8)  #Set overall fig size for document

ts=tsload('rundata.pickle')
ts.name='pH 5 bsa 30nm NPs'
ts.specunit='nm'

ts=ts.sub(ts.darkseries, axis=0)

specplot(ts)
```

```python
ts=ts.as_interval('m') 
areaplot(ts)
```

```python
t1, t2, t3 =ts.ix[:, 10.:14.], ts.ix[:, 30.:35.], ts.ix[:, 47.:50.]
```

```python
base_h20=t1.mean(axis=1)
base_t2=t2.mean(axis=1)
base_t3=t3.mean(axis=1)
```

Now let's look at the absorbance of region 2, setting the baseline (or reference spectrum) to our h20 baseline.  Recall that absorbance is given by $A=-log(\frac{I}{I_o})$

```python
t2.baseline=base_h20
absplot(t2)
```

We can go ahead and plot the post-bsa absorbance spectra over this by setting t3's reference to h20.  Note that because pyuvvis doesn't have any builtin operations to plot two overlapping plots per-se, we just make a generic plot container using pylab and add the labels manually.

```python
t3.baseline=base_h20
x=t3.index

plt.plot(x, t2.as_iunit('a'))
plt.plot(x, t3.as_iunit('a'))

plt.title('Effect of BSA')
plt.xlabel(ts.full_specunit)
plt.ylabel('Absorbance')
```

**This corresponds to a peak shift of 3.9 nm** as shown in code cell below

```python
from pyuvvis.core.utilities import maxmin_xy

t2spr, t3spr= t2.ix[400.:800.], t3.ix[400.:800.]

sprt2=maxmin_xy(t2spr.as_iunit('a'), idx=True, val=False, arg=False).mean()
sprt3=maxmin_xy(t3spr.as_iunit('a'), idx=True, val=False, arg=False).mean()

print 'SPR max in h20: %snm.  Sprtmax after BSA: %snm.  Delta %snm'%(sprt2, sprt3, sprt3-sprt2)
```

It is worth mentioning that I added BSA in a very high concentration, such that only 20uL of volume were added. **As such, nanoparticle dilution was almost negligible, so this signal is mostly due to BSA.**  We can then look at the absorbance of just BSA by setting the reference to that of the bare nanoparticles in h20, but first I'll slice out the short wavelength regions.

```python
t3spr.baseline=t2spr.mean(axis=1)

absplot(t3spr)
```

At some time later, I added more BSA, still in low volume.  This did not cause dilution in the sample.  We can look at the dispersion of this dilution by slicing it as "t4" and referencing it to t3.

```python
t4=ts.ix[:,56.:59.]
t4.baseline=t3.mean(axis=1)
absplot(t4)
```

## PyUvVis Tips and Tricks

Below are some tips to illustrate other faculties of PyUvVis for reference.  They won't add to the above analysis of pyuvvis per-se.

#### Data Representations/Conversions:

We can change between spectral units with much ease using the **specunit** attributes.  This can be done in place, or to generate a new object using the as_specunit() method.  Let's see what spectral units are already builtin.

```python
ts.list_sunits()
```

```python
tnew=ts.as_specunit('f')  #Returns frequency in hZ
```

```python
specplot(tnew)
```

We can change the representation of the intensity of the spectral data (eg full data, abosrbance, transmittance etc...) as long as a baseline has been established.  The syntax and api is much the same as with the specunit.

```python
tnew.list_iunits()
```

```python
tnew.baseline=0
tnew.iunit='%t'
specplot(tnew)
```

Enought metadata is stored so that the representations can always be restored to the original state.  EG:

```python
tnew.iunit=None
tnew.specunit='nm'
specplot(tnew)
```

There is similar functionality for changing temporal represnetations of the data (ie from timestamp to timeranges) but I don't feel like going into that.  Finally, just to show that pyuvvis is primarily a platform for development, I will show how advanced analysis techniques like correlation analysis have already been programmed in (and it was exceedingly easy for that matter).  We will throw in some fancy pyuvvis 3d plot coloring for fun.  (Builtin color maps from here: http://dept.astro.lsa.umich.edu/~msshin/science/code/matplotlib_cm/)

```python
sync_3d(sync, title='Synchronous Spectrum', cmap='gray', contour_cmap='autumn', alpha=0.5)
sync_3d(async, title='Asynchronous Spectrum', c_iso=30, r_iso=30, cmap='summer', contour_cmap='jet', kind='contourf')
```
