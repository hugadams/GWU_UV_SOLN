```python
from pyuvvis import absplot, specplot, areaplot
from pyuvvis.IO.gwu_interfaces import from_oceanoptics

plt.figsize(8,8)  #Set overall fig size for document

ts=from_oceanoptics('Data')
ts.name='12nm BSA'

ts=ts.sub(ts.darkseries, axis=0)
ts.specunit='nm'

specplot(ts)
```

```python
ts=ts.as_interval('m')
areaplot(ts)
```

```python
h20=ts.ix[:,2.0:4.5]
nps=ts.ix[:, 6.0:8.0]
bsa=ts.ix[:, 13.0:15.0]
drop2=ts.ix[:, 20.0::]
```

```python
nps.baseline=h20.mean(axis=1)
bsa.baseline=h20.mean(axis=1)
absplot(nps)
plt.plot(bsa.index, bsa.as_iunit('a'))
plt.xlim(350., 800.)
plt.ylim(0.0,0.4)
```

**This coresponds to a peak shift of 5.41nm** as shown in the code cell below:

```python
from pyuvvis.core.utilities import maxmin_xy

npspr=maxmin_xy(nps.ix[400.0:800.0].as_iunit('a'), idx=True, val=False, arg=False).mean()
bsaspr=maxmin_xy(bsa.ix[400.0:800.0].as_iunit('a'), idx=True, val=False, arg=False).mean()

print 'SPR max in h20: %snm.  Sprtmax after BSA: %snm.  Delta %snm'%(npspr, bsaspr, bsaspr-npspr)
```

Keep in mind there is a negligble dilution effect present here.  See BSA_ph5_notebook for more details on how faint this effect is and how to quantify it.

```python
bsa.iunit=None
nps.iunit=None
bsa.baseline=nps.mean(axis=1)
absplot(bsa.ix[400.0:800.0])
```
