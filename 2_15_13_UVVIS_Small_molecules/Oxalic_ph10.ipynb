```python
from pyuvvis import *
from pyuvvis.IO.gwu_interfaces import from_oceanoptics
plt.figsize(8,8)  #Set overall fig size for document

ts=from_oceanoptics('data')
ts=ts.sub(ts.darkseries, axis=0)
ts.name='ph10_Oxalicacid_test'

specplot(ts)
```

```python
ts=ts.as_interval('m') 
areaplot(ts)
```

```python
areaplot(ts.as_datetime())
```

```python
t1, t2, t3, t4 =ts.ix[:, 1.:5.], ts.ix[:, 17.:22.], ts.ix[:,23.:28.], ts.ix[:, 38.:42.]
```

```python
base_h20=t1.mean(axis=1)
base_t2=t2.mean(axis=1)
base_t3=t3.mean(axis=1)
```

Now let's look at the absorbance of region 2, setting the baseline (or reference spectrum) to our h20 baseline.  Recall that absorbance is given by $A=-log(\frac{I}{I_o})$

```python
t2.baseline=base_h20
absplot(t2)
```

```python
t3.baseline=base_h20
x=t3.index

plt.plot(x, t2.as_iunit('a'))
plt.plot(x, t3.as_iunit('a'))

plt.title('Effect of Oxalic')
plt.xlabel(ts.full_specunit)
plt.ylabel('Absorbance')
```

```python
t4.baseline=base_h20
x=t3.index

plt.plot(x, t2.as_iunit('a'))
plt.plot(x, t4.as_iunit('a'))

plt.title('Effect of Oxalic')
plt.xlabel(ts.full_specunit)
plt.ylabel('Absorbance')
plt.xlim(520,530)
plt.ylim(0.32,0.34)
```

**This corresponds to a peak shift of 0.378 nm** as shown in code cell below

```python
from pyuvvis.core.utilities import maxmin_xy

t2spr, t3spr, t4spr= t2.ix[400.:800.], t3.ix[400.:800.], t4.ix[400.:800.]

sprt2=maxmin_xy(t2spr.as_iunit('a'), idx=True, val=False, arg=False).mean()
sprt4=maxmin_xy(t4spr.as_iunit('a'), idx=True, val=False, arg=False).mean()

print 'SPR max in h20: %snm.  Sprtmax after Oxalic: %snm.  Delta %snm'%(sprt2, sprt4, sprt4-sprt2)
```

```python
t4.baseline=base_t3
plt.plot(t4.index, t4.as_iunit('a'))

plt.title('Effect of the second addition(50ul) of Oxalic')
plt.xlabel(ts.full_specunit)
plt.ylabel('Absorbance')
plt.xlim(400,800)
plt.ylim(-0.015,0.015)
```
