```python
from pyuvvis import *

ts30=tsload('30nm_ph4.3_glutamate.pickle')
ts12=tsload('12nm_ph4.3_glutamate.pickle')
```

# Batch2 30nm nps

```python
ts=ts30.ix[:, ::10] #REDUCE DATASET
ts=ts.sub(ts.darkseries, axis=0)
specplot(ts)
areaplot(ts)
```

**Actually am going to slice raw times based on notes instead of intervals**

```python
h20=ts.ix[:,'2013-02-20 10:34:00':'2013-02-20 10:45:00']
nps=ts.ix[:,'2013-02-20 11:33:00':'2013-02-20 11:41:00']
g1=ts.ix[:,'2013-02-20 11:44:00':'2013-02-20 11:58:00']
g2=ts.ix[:,'2013-02-20 12:15:00':'2013-02-20 12:20:00']

nps.baseline=h20.mean(axis=1)
g1.baseline=h20.mean(axis=1)

nps.iunit='a'
g1.iunit='a'

### Plotting single mean curve, so absplot barfs because of series
plt.plot(nps.index, nps.mean(axis=1), color='g')
plt.plot(nps.index, g1.mean(axis=1), color='r')
plt.xlim(400.0,700.0)
plt.ylim(0.0, 0.35)
```

```python
g1.iunit=None
nps.iunit=None
g1.baseline=nps.mean(axis=1)

nps.iunit='a'
g1.iunit='a'

plt.plot(nps.index, g1.mean(axis=1), color='r')
plt.xlim(250.0,800.0)
plt.ylim(-0.004, 0.002)
```

# How does the dispersion look for the next drop, which should probably be dispersion-free.

```python
g1.iunit=None
g2.baseline=g1.mean(axis=1)

g1.iunit='a'
g2.iunit='a'

p=plt.plot(g2.index, g2.mean(axis=1))
plt.xlim(250.0,800.0)
plt.ylim(-0.008, 0.002)
```

# NOPE THIS APPEARS TO BE ONLY A DILUTION EFFECT!  I CAN INVERT THE LAST PLOT AND SEE IT IS INDEED THE NP SPECTRUM. (INDEED SPR PREAK AT 524.8nm)

```python
plt.plot(g2.index, g2.mean(axis=1))
plt.xlim(250.0,800.0)
plt.ylim(-0.008, 0.002)
plt.gca().invert_yaxis()

print 'spr= %s nm'%round(g2.ix[400.0:700.0].idxmin().mean(), 1)
```
