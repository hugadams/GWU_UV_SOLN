# Absorbance sodium citrate pH 10.

```python
from pyuvvis import *
from pyuvvis.IO.gwu_interfaces import from_oceanoptics

ts=from_oceanoptics('data')
ts.name='Sodium Citrate ph10'
ts.save('NaCit.pickle')
```

```python
ts=ts.sub(ts.darkseries, axis=0)
specplot(ts)
```

```python
areaplot(ts)
ts=ts.as_interval('s')
areaplot(ts)
```

```python
h20=ts.ix[:,0.0:15.0]
nps=ts.ix[:,200.:400.]
citrates=ts.ix[:, 1000.:1400.]

nps.baseline=h20.mean(axis=1)
citrates.baseline=nps.mean(axis=1)
cut=citrates.ix[250.0::]

absplot(nps.ix[400.0::])
absplot(cut)

sprnps=nps.ix[400.0::].as_iunit('a').idxmax().mean(axis=1)
citrates.baseline=h20.mean(axis=1)
sprcit=citrates.as_iunit('a').ix[400.0::].idxmax().mean(axis=1)

r1=lambda x: round(x,1)

print "Near indiscernable peak shift of %s nm between sprnps (%s nm) and spr citrates (%s nm)'"%(r1(sprcit-sprnps), r1(sprnps), r1(sprcit))
```
