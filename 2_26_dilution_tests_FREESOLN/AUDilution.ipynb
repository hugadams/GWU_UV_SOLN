## AuNP 30nm/12nm dilutions

#### Absorbance vs. Concentration data for batch2 30nm/12nm NPs vs. concentration in spectrometer.  Seeing how it compares to the chemistry spectrometer (see dataset 6_19_12).  In this dataset, need to use a line of best fit; whereas on 6-19-12), dilutions were from 85% down to about 0.3% so best fits weren't required.  This analysis offers some unique challenges, since benchtop spectrometer is not capable of strong absorption measurements (eg 100% concentrated np absorbance) so extrapolation is required.

```python
from pyuvvis import *
from pyuvvis.IO.gwu_interfaces import from_oceanoptics

plt.figsize(8,8)

ts=from_oceanoptics('30nm')
```

#### Subtract Baseline and plot data.

```python
ts.sub_base()
specplot(ts)
ts=ts.as_interval('s')
areaplot(ts)
```

#### Slice data into subsets of h20, bare nps and drops.  Convert all subsets into 'absorbance' units for use in rest of analysis.

```python
h20=ts.ix[400.0::, 20.0:60.0]
anps=ts.ix[400.0::, 200.:250.]
d1=ts.ix[400.0::,350.:400.]
d2=ts.ix[400.0::, 500.:530.]
d3=ts.ix[400.0::, 650.:700.]
d4=ts.ix[400.0::,815.::]

allnps={'anps':anps,'d1':d1,'d2':d2,'d3':d3,'d4':d4}

h20base=h20.mean(axis=1)

for dset in allnps.values():
    dset.reference=h20base
    dset.iunit='a'
```

#### Show absorbance plot for all drops.

```python
absplot(anps)
plt.plot(anps.index, d1, color='red')
plt.plot(anps.index, d2, color='green')
plt.plot(anps.index, d3, color='blue')
plt.plot(anps.index, d4, color='purple')
plt.show()
```

#### The starting concentration of np's was 1 part in 6 of 50% nps.  Therefore, it was a 1 part in 12 or 8.3333% dilution.  I can compute the resulting concentrations from the 4  200uL drops as:

```python
vi=3.0 #ml
drops=range(1,5)
cs=[8.33333]
for d in drops:
    cs.append(round(cs[0] * (vi / (vi +.200 * d) ) , 4))

for i,c in enumerate(cs):
    print 'drop %s \t concentration %s %%'%(i,c )
```

#### We can make several abs vs. concentration plots, which should all be linear.  I'll try total area under curve as well as several wavelengths vs. concentration.  To do this, I'll map the area under the curve for each timespectra/drop and merge into a single TimeSpectra object.

```python
areas=dict((k,v.area().mean(axis=1)) for k,v in allnps.items())
short_area=dict((k, v.ix[420.0:450.0].area().mean(axis=1)) for k,v in allnps.items())
peak_area=dict((k,v.ix[522.0:528.0].area().mean(axis=1)) for k,v in allnps.items())
```

#### Store the averaged curves into a new timespectra object and use drop values for columns.

```python
areas=TimeSpectra(areas, index=[0])
peak_area=TimeSpectra(peak_area, index=[0])
short_area=TimeSpectra(short_area, index=[0])


areas.columns=cs
peak_area.columns=cs
short_area.columns=cs
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,8))  #This is quick command to define subplot layout

ax1.plot(cs, areas.transpose(), color='black')
ax1.scatter(cs, areas.transpose(), color='black')
ax2.plot(cs, short_area.transpose(), color='red')
ax2.scatter(cs, short_area.transpose(), color='red')
ax3.scatter(cs, peak_area.transpose(), color='blue')
ax3.plot(cs, peak_area.transpose(), color='blue')

ax1.set_xlabel('Drop Concentration %')
ax2.set_xlabel('Drop Concentration %')
ax3.set_xlabel('Drop Concentration %')

ax1.set_ylabel('Absorbance (area under the curve)')

ax1.set_title('Total Area dilution')
ax2.set_title('ShortWavelength Area dilution')
ax3.set_title('PeakWavelength Area dilution')

```

##### We see that sampling various wavelength regions doensnt make a huge difference.  Also, this change is not quite perfectly linear, but this is mostly due to drop issues/pipetter issues.  I know the relationship should be linear.

## Perhaps using the average absorbance and not the area under the curve is what I meant to do, even though the area is in-fact constant over the range, which is good to know.

```python
avgs=dict((k,v.mean(axis=1).mean(axis=1)) for k,v in allnps.items())
short_avgs=dict((k, v.ix[420.0:450.0].mean(axis=1).mean()) for k,v in allnps.items())
peak_avgs=dict((k,v.ix[522.0:528.0].mean(axis=1).mean()) for k,v in allnps.items())
```

```python
avgs=TimeSpectra(avgs, index=[0])
peak_avgs=TimeSpectra(peak_avgs, index=[0])
short_avgs=TimeSpectra(short_avgs, index=[0])

avgs.columns=cs
peak_avgs.columns=cs
short_avgs.columns=cs
```

## Fit various area data using linear fit.

There are several fitting/polynomical packages.  For general fitting scipy has curve_fitting and optimization utilities.  Numpy has an advanced packagd called numpy.polynomical.polynomial.  See below for an example of how to use it.  Althought the polyval function evaluates a polynomial possibly more generally, the poly1d seems to be a nice class to store a polynomial and return data/attributes of it.

http://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html#numpy.poly1d

```python
from numpy.polynomial.polynomial import polyfit, polyval
from numpy import poly1d

r2 = lambda x: round(x,2)

cs_full_range=range(1,100) #Extrapolating from 0-100%
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,8))  #This is quick command to define subplot layout


#Area sources
#sources=((ax1,areas,'black'),(ax2,short_area,'red'), (ax3,peak_area,'blue'))

#AVG Sources
sources=((ax1,avgs,'black'),(ax2,short_avgs,'red'), (ax3,peak_avgs,'blue'))

x_drops=[float(i) for i in range(1,100)]

print '%s \t %s \t %s \t %s'%('color', 'Absorbance at 100%', 'slope','intercept')

for ax,source, color in sources:
    b, m=polyfit(source.columns, source.transpose(), 1)  #Returns lower power first (NUMPY BUG)
    p=poly1d([m[0], b[0]]) #Takes in list as decreasing powers
    ax.plot(x_drops, p(x_drops), color=color )
    ax.scatter(cs, source.transpose(), color=color)
    ax.set_xlim(0.0,100.0)
    ax.set_xlabel('Drop Concentration %')
    ax.set_ylim(0.0)
    
    print '%s \t %s \t %s \t %s'%(color, p(100), m[0], b[0])
    
ax1.set_ylabel('Absorbance (average)')
ax1.set_title('Average Absorbance vs dilution')
ax2.set_title('Short Wavelength Absorbance vs dilution')
ax3.set_title('Peak Wavelength Absorbance vs dilution')

```

## At sprmax, we find the absorbance at 100% to be 3.412 Absorbance units

We can map our plot of the relative concentration ($C_r$) to that of the real concentration ($C_t$) via:

$A=mC_r$  

$A=\epsilon(r) C_t$

$C_t = \frac{m}{\epsilon(r)} \times C_t$

From our plot above, $m=0.03416$  and from an estimate in a paper, let's choose $\epsilon=2.93\times 10^9$.  (This is estimated for citrate capped particles with R=25nm, but should see my notes on the subject before being 100% behind this figure.  Note, epsilon is a scalar because generally people measure it for the SPR max.



```python
m=0.03416
eps=2.93*10**9
ctrue=(m/eps) * anps.columns.values
ctrue
```
