# Dropwise mixture concentration calculator

### RUN PARAMETERS (CHANGE HERE):    

```python
VO = 3          # Volume of solvent in cuvette: (mL)
SOLVENT = 'h20' # Liquid in cuvette before pumping

VD = 0.2        # Volume of drops being pumped in/out: (mL)
SOLUTE = 'PBS' #Liquid being pumped in

# Yo, x array, interesting concentrations: best to not change
CI = 1.0
MAX_DROPS = 100
interesting_cs = [0.8, 0.9, 0.95, 0.99] #Concentrations to draw vlines
```

### Generate an array of xvalues (eg drops/steps) and compute the following exponential

$C_f = C_i \times  e^{(- N \frac{V_d}{V_o})}$

```python
drops = np.arange(MAX_DROPS) 
CF = CI * np.exp(-drops * (VD/VO) )
```

```python
def time_constant(conc):
    ''' Given a concentration, returns correspding x value.'''
    return (-VO/VD) * log(1.0 - conc)
```

### Plot of concentration vs. Number drops

```python
plt.plot(drops, CF, color='blue', lw=3)
plt.plot(drops, 1.0 - CF, color='red', lw=3)

plt.title('C vs. Steps')
plt.ylabel('Concentration')
plt.xlabel('Steps/Drops')
plt.legend([SOLVENT, SOLUTE])
plt.grid()
plt.minorticks_on()

# Plot some important timepoints
vlines = [time_constant(item) for item in interesting_cs]
plt.vlines(vlines, 0, 1.0, linestyles='--', lw=3)

# Shorthand rounding functions
r2 = lambda x: str(round(x, 2))
r1 = lambda x: str(round(x, 1))

for idx, xvalue in enumerate(vlines):
    plt.text(xvalue + 1, 0.5, "C = %s\nX = %s" % (r2(interesting_cs[idx]), r1(xvalue) ), size=12 )
```

# Plot of concentration vs. total volume of solute/buffer consumed

```python
volume_used = drops * VD

plt.plot(volume_used, CF, color='blue', lw=3)
plt.plot(volume_used, 1.0 - CF, color='red', lw=3)

plt.title('C vs. Total Volume')
plt.ylabel('Concentration')
plt.xlabel('Volume of solute pumped (mL)')
plt.legend([SOLVENT, SOLUTE])
plt.grid()
plt.minorticks_on()

vlines = [time_constant(item)*VD for item in interesting_cs]
plt.vlines(vlines, 0, 1.0, linestyles='--', lw=3)

for idx, xvalue in enumerate(vlines):
    plt.text(xvalue + 1*VD, 0.5, "C = %s\nX = %s" % (r2(interesting_cs[idx]), r1(xvalue) ), size=12 )
```
