## Binding rates in H20 VS PBS.

```python
from pyuvvis import tsload, areaplot, TimeSpectra
plt.figsize(8,8)

f1,f2,f4=tsload('f1_bsa.pickle'), tsload('f2_bsa.pickle'), tsload('f4_bsa_pbs.pickle')
```

```python
fibs=(f1,f2,f4)
refunit='r'

### Should work in a loop but seems to be a bug
f1=f1.as_interval('m')
f2=f2.as_interval('m')
f4=f4.as_interval('m')

f1.reference=0
f2.reference=0
f4.reference=0

f1.iunit=refunit
f2.iunit=refunit
f4.iunit=refunit

f1=f1.ix[440.0:680.0]
f2=f2.ix[440.0:680.0]
f4=f4.ix[440.0:680.0]

for f in fibs:
    f.baseline.index=f.index #Current bug
    f.sub_base()
```

#### Handle f3 separately as it had several separate events- will extract bsa portion for now

```python
f3=tsload('f3_bsa_dtssp.pickle')
```

```python
f3=f3.as_interval('m')
f3=f3.ix[440.0:680.0, 66.3::]
f3.reference=0
f3.iunit=refunit
f3.columns=f3.columns.values-66.3
```

```python
### Crop and align f1/f4 to fit f2 data
f1=f1.ix[:, 16.1::]  
f1.columns=f1.columns.values-16.1  #NEED OFFSET FUNCTION

f2=f2.ix[:,.3::]
f2.columns=f2.columns.values-.3

f4=f4.ix[:,1.2::]
f4.columns=f4.columns.values-1.2
```

```python
areaplot(f1, color='green', grid=True)
plt.plot(f2.columns, f2.area().transpose(), lw=3, color='red')
plt.plot(f3.columns, f3.area().transpose(), lw=3, color='orange')
plt.plot(f4.columns, f4.area().transpose(), lw=3, color='black')
plt.xlim(0.0,15.0)
plt.xlabel('Minutes')
plt.title('BSA Binding in various media')
plt.legend(['F1:H20', 'F2:PBS-DTSSP', 'F3: PBS-DTSSP With Quenching', 'F4:PBS'], loc='lower right')
```

So, unless this is a fluke, quenching actually in this cause not only did not stop binding, but possibly enhanced it over functionaly DTSSP.  It may be that Quenched DTSSP helps prevent thiol binding; however, it's tough to say without another trial or two.
